```python
import math
import random
import statistics as st
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def flip_coin():
    return random.randint(0,1)
```

```python
def flip_ALOT():
    heads_list = []
    tails_list = []
    for i in range(17):
        heads = 0
        tails = 0
        for j in range(2**i):
            flip = flip_coin()
            if flip == 1:
                heads += 1
            else:
                tails += 1

        heads_list.append(heads)
        tails_list.append(tails)
                        
    return heads_list, tails_list

heads_list, tails_list = flip_ALOT()

plt.plot(heads_list)
# plt.xscale('log')
plt.plot(tails_list)
# plt.xscale('log')
plt.show()
```

```python
diff = []
ratio = []
max = len(heads_list)
min = 4
for i in range(min, max):
    diff.append((heads_list[i] - tails_list[i]))

for i in range(min, max):
    ratio.append((heads_list[i] / tails_list[i]))
                
print(diff)

plt.plot(diff)
plt.plot(ratio)
plt.show()

```

```python
diff = []
at_list = []
ratio = []

max = len(heads_list)
min = 4
for i in range(min, max):
    diff.append((heads_list[i] - tails_list[i]))
    at_list.append(2**i)

for i in range(min, max):
    ratio.append((heads_list[i] / tails_list[i]))
                
print(diff)

plt.plot(diff)
plt.xscale('log')
plt.show()

plt.scatter(diff, at_list)
# plt.xscale('log')
plt.xlabel("Diff b/t Heads & Tails")
plt.ylabel("Number of Flips")
plt.title("Coin Flips")
plt.show()

plt.scatter(ratio, at_list)
plt.xlabel("PERCENTAGE Diff b/t Heads & Tails")
plt.ylabel("Number of Flips")
plt.title("Coin Flips")
plt.show()

```

```python
# do it 20 times
def get_flip_round():
    heads = 0
    tails = 0
    for i in range(2**16):
        flip = flip_coin()
        if flip == 1:
            heads += 1
        else:
            tails += 1
                        
    return heads, tails

heads_list = []
tails_list = []
for each in range(20):
    heads, tails = get_flip_round()
    
    heads_list.append(heads)
    tails_list.append(tails)

plt.ylabel("Standard Deviation")
plt.xlabel("Mean")    
plt.hist(heads_list)
mean = st.mean(heads_list)
stdev = st.stdev(heads_list)
ymin, ymax = plt.ylim()

plt.vlines(mean, ymin, ymax, color='red')
plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, color='red', linestyles='dashed')
plt.show()

plt.ylabel("Standard Deviation")
plt.xlabel("Mean")
plt.hist(tails_list)
mean = st.mean(tails_list)
stdev = st.stdev(tails_list)
ymin, ymax = plt.ylim()

plt.vlines(mean, ymin, ymax, color='red')
plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, color='red', linestyles='dashed')
plt.show()

```

```python
#this keeps adding to the list - to get close to "1.000"
heads_list = []
tails_list = []
ration = []

for each in range(20):
    heads, tails = get_flip_round()
    
    heads_list.append(heads)
    tails_list.append(tails)

for i in range(len(heads_list)):
    ratio.append(heads_list[i] / tails_list[i])
    
total_ratio = sum(ratio) / len(ratio)

print(total_ratio)

plt.plot(ratio)
plt.show()
    
```
